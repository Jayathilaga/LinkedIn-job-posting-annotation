```python
import pandas as pd
```

```python
df=pd.read_excel('hari_annotated.xlsx')
```

```python
df=df.drop(df.columns[:-3],axis=1)
df['skills'] = df['skills'].str.replace(' ', '')
for i in range(df.shape[0]):
    if pd.isna(df['skills'][i]):
        df['skills'][i]='na'
    df['skills'][i]=df['skills'][i].lower().split(',')
```

```python
df
```

```python
df=df.rename(columns={'class name':'class_name','expereince':'experience'})
```

```python
df.loc[df['class_name'] =='Data Science', 'class_name'] = 'data_scientist'
df.loc[df['class_name'] =='Machine Learning Engineer', 'class_name'] = 'machine_learning_engineer'
df.loc[df['class_name'] =='NLP', 'class_name'] = 'nlp_engineer'
```

```python
df
```

```python
df.to_csv('hari_annotated.csv',index=False)
```
